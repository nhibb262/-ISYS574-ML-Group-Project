```python
from google.colab import drive
drive.mount('/content/drive')

# Create project folder in Drive
import os
PROJECT_PATH = '/content/drive/MyDrive/sf-events-explorer'
os.makedirs(f'{PROJECT_PATH}/data/raw', exist_ok=True)
os.makedirs(f'{PROJECT_PATH}/data/processed', exist_ok=True)
os.makedirs(f'{PROJECT_PATH}/models', exist_ok=True)
```

```python
import pandas as pd

# Direct download from SF Open Data
DATA_URL = "https://data.sfgov.org/api/views/8i3s-ih2a/rows.csv?accessType=DOWNLOAD"

# Load into pandas
df_raw = pd.read_csv(DATA_URL)
print(f"Downloaded {len(df_raw)} records")
print(f"Columns: {list(df_raw.columns)}")

# Save to Google Drive
df_raw.to_csv(f'{PROJECT_PATH}/data/raw/Our415_Events_and_Activities.csv', index=False)
print(f"Saved to {PROJECT_PATH}/data/raw/")
```
